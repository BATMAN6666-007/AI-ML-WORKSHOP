```python
a=np.array([1,2,3])
b=np.array([4,5,6])

sum_array=a+b
sub_array=a-b
mul_array=a*b
div_array=a/b
exp_array=np.exp(a)
log_array=np.log(a)
sqrt_array=np.sqrt(a)

print(sum_array)
print(sub_array)
print(mul_array)
print(div_array)
print(exp_array)
print(log_array)
print(sqrt_array)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
mean = np.mean(arr)
sum_row=np.sum(arr,axis=0)
max_cols=np.max(arr,axis=0)
variance=np.var(arr)
print(mean,sum_row,max_cols,variance)
```
