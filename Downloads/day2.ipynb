```python
q1,q2,q3 = np.percentile(df['fare'],[25,50,75])
print(q1,q2,q3)
IQR=q3-q1
print(IQR)
lower_bound=q1-(1.5*IQR)
upper_bound=q3+(1.5*IQR)
print(lower_bound,upper_bound)

outlier =[]
for i in df['fare']:
    if i<lower_bound or i>upper_bound:
        outlier.append(i)
print(outlier)
```
